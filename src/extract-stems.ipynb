```python
import shutil
import os

from pydub import AudioSegment
import stempeg
```

```python
def extract_stems(stem_file, output_dir):

  data, sample_rate = stempeg.read_stems(stem_file)
  num_stems = len(data)


  for i in range(num_stems):
    stem_name = f"stem_{i+1}.wav" 
    output_path = f"{output_dir}/{stem_name}"
    stempeg.write_audio(output_path, data[i], sample_rate, codec="pcm_s16le")

  print(f"Successfully extracted {num_stems} stems to {output_dir}")
```
